```python
player_salary=fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False) #highest salary list
player_salary.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=['aqua','green','red','orange','blue'])
plt.show()
```
